### 1.2 Part A
<a id="problem1.2a"></a>
<a id="#problem1.2a"></a>

$h(\mathbf{x}) = \text{sign}(\mathbf{W}^T\mathbf{X})$ is undefined when $\mathbf{W}^T\mathbf{X} = 0$ since 0 is neither positive or negative. Hence, this is the boundary between $h(\mathbf{x}) = 1$ and $h(\mathbf{x}) = -1$

We can show the boundary is a line as opposed to any other type of squiggly boundary threw the inequality...

$$h(\mathbf{x}) = 1 \implies \mathbf{W}^T\mathbf{X} > 0   \tag{1}$$ 

$$ \mathbf{W}^T \cdot \mathbf{X} > 0 = \tag{2}$$

$$ w_0 + w_1x_1 + w_2x_2 > 0 \tag{dot product}$$

$$ x_2 > \frac{-w_1}{w_2}x_1 - \frac{w_0}{w_2}\tag{solve for x2}$$

In the $x_1$-$x_2$ plane, this inequality is a line because $x_2$ and $x_1$ differ by a linear factor (plus a constant). This shows that the regions where $h(x) = 1$ and $h(x) = -1$ are separated by a line.

````python
# By Andres Columna
# Inspired by the textbook


import numpy as np
from numpy import sign, dot
from random import choice, shuffle, randint
import matplotlib.pyplot as plt

def train_perceptron(training_data, step_size = 1):
    '''
    Train a perceptron model given a set of training data
    :param training_data: A list of data points, where training_data[0]
    contains the data points and training_data[1] contains the labels.
    Labels are +1/-1.
    :return: learned model vector
    '''
    X = training_data[0]
    y = training_data[1]
    model_size = X.shape[1]
    num_of_samples = X.shape[0]
    w = np.zeros(model_size)  # np.random.rand(model_size)
    iteration = 1
    
    while True:

        # pick random row
        shuffled_rows = list(range(num_of_samples))
        shuffle(shuffled_rows) #in-place
        # if all rows in the ordering don't need updating of W-vector, algorithm
        # has converged
        iterations_without_correction = 0
        
        # Pick one misclassified example.
        for row_index in shuffled_rows:
            x, y_t = X[row_index], y[row_index]

            signal_t = sign(dot(w, x))

           
            if signal_t * y_t <= 0:
                w = w + (y_t * x)
                iteration += 1 # an iteration is equivalent to a correction
            else:
                iterations_without_correction += 1
                continue

            if iteration > 500:
                print("an infinite loop likely happened!")
                return -1

        else:
            # Check the convergence criteria
            if iterations_without_correction == num_of_samples:
                print("Iterations: ", iteration)
                return w

        iteration += 1
    
    return w

def print_prediction(model, data, rnd_y):
    '''
    Print the predictions given the dataset and the learned model.
    :param model: model vector
    :param data:  data points
    :return: nothing
    '''
    result = np.matmul(data, model)
    predictions = np.sign(result)
    for i in range(len(data)):
        print("{}: {} -> {}, actual: {}".format(data[i][:3], result[i], predictions[i], rnd_y[i]))


if __name__ == '__main__':
    rnd_x = np.array([[0, 1, 1], \
                      [0.6, 0.6, 1], \
                      [1, 0, 1], \
                      [1, 1, 1], \
                      [0.3, 0.4, 1], \
                      [0.2, 0.3, 1], \
                      [0.1, 0.4, 1], \
                      [0.5, -0.1, 1]])

    rnd_y = np.array([1, 1, 1, 1, -1, -1, -1, -1])
    rnd_data = [rnd_x, rnd_y]
    trained_model = train_perceptron(rnd_data)
    print("Model:", trained_model)
    print_prediction(trained_model, rnd_x, rnd_y)
````
